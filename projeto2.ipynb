```python
class Player(object):
    cards = [0]
    bet = 0
    score = 0
    def __init__(self, name, bet=0):
        self.nome = name
        
        self.bet = bet
    
    def cards(self, cards):
        self.cards = cards
        
    def setBet(self, bet):
        self.bet = bet
        
    def getBet(self):
        return self.bet
    def getCards(self):
        return self.cards
    
    def setPlayerScore(self, score):
        self.score += score
    def getPlayerScore(self):
        return self.score
        
    def showCards(self):
        print("----------player's cards----------")
        for i in self.cards:
            
            print(i)
        print("------------------------------------")
    def showPlayerScore(self):
        
        print("**********score************")
        print(self.score)
        print("********************************")
        
   
    
    
```

```python
class Dealer(Player):
    def showCards(self):
        print("----------dealer's cards----------")
        for i in self.cards:
            print(i)
            break
            
        print("------------------------------------")
    def showAllCards(self):
        print("----------All dealer's cards----------")
        for i in self.cards:
            print(i)

        print("------------------------------------")
```

```python
class Card(object):
    
    def __init__(self, name, value, symbol):
        self.name = name
        self.value = value
        self.symbol = symbol
    def getName(self):
        return self.name
    def getValue(self):
        return self.value
    
    def __str__(self):
        return "Name: %s , Value: %s, Sybol: %s " %(self.name, self.value, self.symbol)
```

```python
import random
class Deck(object):
    c = []
    def deckCreate(self):
        cards = []
        value = 0
        for i in ('A','K','Q','J',10,9,8,7,6,5,4,3,3):
            if type(i) != int:
                value = 10
            else:
                value = i
           
            card = (Card(str(i), value, '♦'))
            cards.append(card)
            
            card = (Card(str(i), value, '♣'))
            cards.append(card)
            
            card = (Card(str(i), value, '♥'))
            cards.append(card)
            
            card = (Card(str(i), value, '♠'))
            cards.append(card)
            
            random.shuffle(cards)
        return cards
            
    
```

```python


player = Player("John")
playerBet = input("Digite o valor da Aposta: ")
player.setBet(playerBet)
print(player.getBet())
dealer = Dealer("Dealer")

#create deck shuffled
deck = Deck()
deckCards =  deck.deckCreate()

#list to get two cards of the deck
gotCards = []
gotCardsDealer = []
gotCards.append(deckCards[0])
player.setPlayerScore(deckCards[0].getValue())
gotCards.append(deckCards[1])
player.setPlayerScore(deckCards[1].getValue())


gotCardsDealer.append(deckCards[2])
dealer.setPlayerScore(deckCards[2].getValue())
gotCardsDealer.append(deckCards[3])
dealer.setPlayerScore(deckCards[3].getValue())


#take off two cards of deck
del(deckCards[0])
del(deckCards[1])
del(deckCards[2])
del(deckCards[3])

#cards to player
player.cards(gotCards) 
dealer.cards(gotCardsDealer)

player.showCards()
player.showPlayerScore()
print("----------dealer-------------")
dealer.showCards()
```
